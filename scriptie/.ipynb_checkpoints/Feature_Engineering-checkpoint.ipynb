```python
def plot_multiclass_roc(X_test, y_test, y_score, n_classes, labelEncoder=None, figsize=(17, 6), title=None):
    # structures
    fpr = dict()
    tpr = dict()
    roc_auc = dict()

    # calculate dummies once
    y_test_dummies = pd.get_dummies(y_test, drop_first=False).values
    for i in range(n_classes):
        fpr[i], tpr[i], _ = roc_curve(y_test_dummies[:, i], y_score[:, i])
        roc_auc[i] = auc(fpr[i], tpr[i])

    labelEncoder
    # roc for each class
    fig, ax = plt.subplots(figsize=figsize)
    ax.plot([0, 1], [0, 1], 'k--')
    ax.set_xlim([0.0, 1.0])
    ax.set_ylim([0.0, 1.05])
    ax.set_xlabel('False Positive Rate')
    ax.set_ylabel('True Positive Rate')
    if title:
        ax.set_title(title)
    else:
        ax.set_title('ROC AUC plot')
    
    for i in range(n_classes):
        if labelEncoder:
            label = labelEncoder.inverse_transform([i])
            ax.plot(fpr[i], tpr[i], label='ROC curve (area = {}) for label {}'.format(roc_auc[i], label))
        else:
            ax.plot(fpr[i], tpr[i], label='ROC curve (area = %0.2f) for label %' % (roc_auc[i], i))
    ax.legend(loc="best")
    ax.grid(alpha=.4)
    sb.despine()
    plt.show()
    return fig
        
def plot_feature_importance(importance, names, title, savename):
    #Create arrays from feature importance and feature names
    feature_importance = np.array(importance)
    feature_names = np.array(names)

    #Create a DataFrame using a Dictionary
    data={'feature_names':feature_names,'feature_importance':feature_importance}
    fi_df = pd.DataFrame.from_dict(data)

    #Sort the DataFrame in order decreasing feature importance
    fi_df.sort_values(by=['feature_importance'], ascending=False,inplace=True)

    fi_df['feature_names'] = fi_df['feature_names'].astype('category')
    display(fi_df.info())
    
    fig = px.bar(fi_df,
                 title='Chi-Squaredtest top 25',
                 x='feature_importance',
                 y='feature_names',
                 color='feature_names',
                 color_discrete_sequence=px.colors.qualitative.Dark24,
                 width=1500, height=1000)
    fig.update_xaxes(title='Chi-Square')
    fig.update_yaxes(title='Feature Names')
    
    fig.update_layout(
    font_family="Cambria",
    title_font_size =25,
    title_font_family="Cambria",
    title_x=0.5)
    fig.show()
```

```python
# feature selection
def select_features(features, target, score_func='chi2', k=25):
    fs = SelectKBest(score_func=chi2, k=k)
    fs.fit(features, target)
    return fs

def plot_ROC_curve(model, xtrain, ytrain, xtest, ytest, modelname):
    # Creating visualization with the readable labels
    visualizer = ROCAUC(model, encoder={0: 'Dodelijk', 
                                        1: 'Letsel', 
                                        2: 'Uitsluitend materiele schade'}, size=(1080, 720))
                                        
    # Fitting to the training data first then scoring with the test data                                    
    visualizer.fit(xtrain, ytrain)
    visualizer.score(xtest, ytest)
    name_file = 'ROC' + modelname
    visualizer.show(outpath=name_file)
    
    return visualizer
```

```python
# Split the data in features and target
features = df_TOTAAL.loc[:, df_TOTAAL.columns != 'AP3_CODE']
target = df_TOTAAL.loc[:, 'AP3_CODE']
target_names = target.unique()

# Encode targets to labels
target_enc = LabelEncoder()
y = target_enc.fit_transform(target)
```

### Pijplein voor Encoden van Categorische Features

```python
ordinal_categorical_features = ['WDK_ID','WVG_ID','MAXSNELHD','LGD_ID', 'WGD_CODE_1']
ordinal_categorical_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='constant', fill_value='Onbekend')),
    ('ordinalEnco', OrdinalEncoder())])

nominal_categorical_features = [col for col in features if col not in ordinal_categorical_features]
print(nominal_categorical_features)
nominal_categorical_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='constant', fill_value='Onbekend')),
    ('nominalEnco', OneHotEncoder(handle_unknown='ignore'))])

preprocessor = ColumnTransformer(
    transformers=[
#         ('numerical_minmax', numerical_percentage_tranformer, numerical_percentage_features),
#         ('numerical_standard', numerical_value_transformer, numerical_value_features),
        ('categorical_nominal', nominal_categorical_transformer, nominal_categorical_features),
        ('categorical_ordinal', ordinal_categorical_transformer, ordinal_categorical_features)
    ])
preprocessor
```

```python
# get Feature names from onehotencoder
feature_selection = preprocessor.fit_transform(features)

display(feature_selection.shape)
X_train, X_test, y_train, y_test = train_test_split(feature_selection, y, stratify=y, random_state=42, train_size=.7)
onehot_featurernames = preprocessor.transformers_[0][1][1].get_feature_names(nominal_categorical_features)

# # Get feature names
feature_names = np.concatenate((ordinal_categorical_features, onehot_featurernames))

# # # filter out Onbekend
mask = np.array([False if "Onbekend" in name else True for name in feature_names])
feature_names = feature_names[mask]
feature_selection = feature_selection[:, mask]

# feature_names, feature_selection

```

```python
# Chi^2 anlayse
k = 25
fs = select_features(feature_selection, target, score_func='chi2', k=k)

mask = fs.get_support().tolist()
K_feature_names = feature_names[mask]

# len(fs.scores_)
plot_feature_importance(fs.scores_[mask], K_feature_names, 'Chi-Squared feature importance', 'chi2')
```

```python
X2 = chi2(feature_selection, target)[0].tolist()
p_value = chi2(feature_selection, target)[1].tolist()
data = {'feature_names':feature_names.tolist(), 'chi_Squared': X2, 'p_value':p_value}
chi2_df = pd.DataFrame.from_dict(data)
chi2_df
chi2_df.sort_values(by='chi_Squared', ascending=False)[:50].to_excel("chi2(50).xlsx")
```

```python
# features.astype(str)
catboost_features = features.applymap(str)
X_train, X_test, y_train, y_test = train_test_split(catboost_features, y, stratify=y, train_size=.7)



cat_features = catboost_features.columns
class_weight_train = compute_sample_weight('balanced', y=y_train)
class_weight_test = compute_sample_weight('balanced', y=y_test)

train_dataset = Pool(data=X_train,
                     label=y_train,
                     cat_features=cat_features,
                     weight=class_weight_train)

test_dataset = Pool(data=X_test,
                    label=y_test,
                    cat_features=cat_features,
                    weight=class_weight_test)
```
